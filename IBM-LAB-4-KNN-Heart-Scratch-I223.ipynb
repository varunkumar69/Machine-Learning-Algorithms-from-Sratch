```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
df.shape
```

```python
#classifiying the data based on the target variable
target1=[]
target0=[]


for i in range(len(df)):
    if df.iloc[i][-1]==1:
        target1.append(list(df.iloc[i][0:-1]))
    else:
        target0.append(list(df.iloc[i][0:-1]))
```

```python
#assigning respective values to its key
data={1:target1,0:target0}
data
```

```python
#pred_pt=[63.0, 1.0, 3.0, 145.0, 233.0, 1.0, 0.0, 150.0, 0.0, 2.3, 0.0, 0.0, 1.0]
distances=[]
def KNN(data,pred_pt,k):
    distances = []
    for grp in data:
        #print(grp)
        for pt in data[grp]:
            dist = np.linalg.norm(np.array(pred_pt)-np.array(pt))
            distances.append((dist,grp))
    print(sorted(distances)[0:k])
    
    votes = [i[1] for i in sorted(distances)[0:k]]
    
    print(votes)
    return Counter(votes).most_common()[0][0]
```

```python
pred_pt=[57.0, 1.0, 0.0, 165.0, 289.0, 1.0, 0.0, 124.0, 0.0, 1.0, 1.0, 3.0, 3.0]
k=4
Class=KNN(data,pred_pt,k)
print("Class : ",Class)
```
