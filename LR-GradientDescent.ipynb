```python
import numpy as np
def predict(x,parameters):
  return x*parameters[1]+parameters[0]
```

```python
#finding value of m and c using gradient descent

import matplotlib.pyplot as plt
def update_parameter(x,y,parameter,alpha,epoch): # we will calculate new values of m and c
    sum_errors=list()
    for no_time in range(epoch):
        sum_error=0
        for i in range(len(x)):
            y_hat=predict(x[i],parameter)
            error=y_hat-y[i]
            sum_error=sum_error+error**2
            #sum_error=sum_error+abs(error)
            parameter[1]=parameter[1]-(alpha*error*x[i]) # for  new value of m
            parameter[0]=parameter[0]-(alpha*error)# for new value of c
        sum_errors.append(sum_error)
        print("epoch ",no_time,"sum error : ",sum_error)
    plt.plot(range(epoch),sum_errors)
    plt.xlabel("Epoch")
    plt.ylabel("Sum error")
    plt.show()
    
        
        
    print("m= ",parameter[1],"c= ",parameter[0])
    return parameter
        
```

```python
x=np.array([89,66,78,111,44,77,80,66,109,75,85,110,45,79,90,76,81])
y=np.array([7,5.4,6.6,7.4,4.8,6.4,7,5.6,7.3,6.4,6.9,7.3,4.8,6.8,7,6.5,7])
#x=[1,2,3,4,5]
#y=[1,2,3,4,5]
parameter=[0,0]#[c,m]
alpha=0.0001 #learning rate,more learning rate more time it will take(learning rate can be 0.1 or 0.01 or 0.001)
epoch=500 #more the epoch ,more the accuracy
up_parameter=update_parameter(x,y,parameter,alpha,epoch)
print("updated parameters",up_parameter)
```
